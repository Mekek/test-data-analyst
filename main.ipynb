# Тестовое задание
 Это продажи по сети Магнит. Нужно их соединить в один. Цифры ( столбцы 'Продажи в шт.', 'Себестоимсть в руб.', 'Продажи в руб.') суммируем если идет совпадение в столбцах "Код" + "Код позиции".

 Для начала преобразуем файлы или xlsb в csv.

```python
import pandas as pd
from pyxlsb import open_workbook
import os

def xlsb_to_csv(input_file: str, output_file: str):
    with open_workbook(input_file) as wb:
        sheet_names = wb.sheets
        first_sheet = sheet_names[0]  
        
        df = []
        with wb.get_sheet(first_sheet) as sheet:
            for row in sheet.rows():
                df.append([item.v for item in row])
        
        df = pd.DataFrame(df)
        df.to_csv(output_file, index=False, header=False, encoding='utf-8')
    print(f"Файл {input_file} преобразован в {output_file}")


input_path_1 = "input_data/Магнит Уход и гигиена 07.24.0.xlsb"
input_path_2 = "input_data/Магнит Уход и гигиена 08.24.0.xlsb"
output_path_1 = "input_data/7_24_0.csv"
output_path_2 = "input_data/8_24_0.csv"

if os.path.exists(input_path_1):
    xlsb_to_csv(input_path_1, output_path_1)
else:
    print(f"Файл {input_path_1} не найден.")

if os.path.exists(input_path_2):
    xlsb_to_csv(input_path_2, output_path_2)
else:
    print(f"Файл {input_path_2} не найден.")

```

```python
df_7_24_0 = pd.read_csv(output_path_1)
df_8_24_0 = pd.read_csv(output_path_2)
```

В файле "Магнит Уход и гигиена 07.24.0.xlsb" (df_7_24_0) нет столбца "Уровень.4", поэтому добавляем этот столбец и заполняем пустыми значениями, чтобы объединение прошло корректно.

```python
if "Уровень.4" not in df_7_24_0.columns:
    df_7_24_0["Уровень.4"] = pd.NA 
```

```python
# Объединяем два датафрейма
df_merged = pd.concat([df_7_24_0, df_8_24_0], ignore_index=True)

print(df_merged.size)
# Группируем по "Код" и "Код позиции", суммируя числовые колонки
df_result = df_merged.groupby(["Код", "Код позиции"], as_index=False).agg({
    "Продажи в шт.": "sum",
    "Себестоимсть в руб.": "sum",
    "Продажи в руб.": "sum",
    **{col: "first" for col in df_merged.columns if col not in ["Код", "Код позиции", "Продажи в шт.", "Себестоимсть в руб.", "Продажи в руб."]}
})
```

При попытке сохранить результаты на одном листе получаем ошибку из-за слишком больших размеров, поэтому сохраняем данные в одном файле, но на разных листах.

```python
try:
    # Пытаемся сохранить весь датафрейм в один лист
    df_result.to_excel("output_data/финальный_отчет.xlsx", index=False, engine="openpyxl")
except ValueError as e:
    # Если ошибка из-за размера, разбиваем на части
    print(f"Ошибка: {e}. Сохраняем в несколько листов.")

    chunk_size = 1_000_000  # Количество строк на один лист
    chunks = [df_result.iloc[i:i + chunk_size] for i in range(0, len(df_result), chunk_size)]

    with pd.ExcelWriter("output_data/финальный_отчет.xlsx", engine="openpyxl") as writer:
        for i, chunk in enumerate(chunks):
            chunk.to_excel(writer, sheet_name=f"Лист_{i+1}", index=False)

    print("Файл успешно сохранен в нескольких листах.")
```
